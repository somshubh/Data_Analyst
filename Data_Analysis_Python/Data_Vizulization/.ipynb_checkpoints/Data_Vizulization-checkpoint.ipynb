```python
import pandas as pd
import matplotlib.pyplot as py
import seaborn as sns
sns.set()
```

```python
df_car_data = pd.read_csv("bar_chart_data.csv")
```

```python
df_car_data
```

```python
py.Figure(figsize=(9,6))
```

```python
py.bar(x=df_car_data['Brand'],
       height = df_car_data['Cars Listings'],
       color = "midnightblue")
py.title("Car Listing By Brand", fontsize = 15, fontweight = 'bold')
py.ylabel('Number of Listing', fontsize = 13)
py.yticks(fontsize = 13)
py.xticks(rotation = 45)
py.show()
py.savefig('Car_List.png')
```

```python
df_car_data_pie = pd.read_csv("pie_chart_data.csv")
```

```python
df_car_data_pie
```

```python
py.Figure(figsize=(10,8))
py.pie(df_car_data_pie['Number of Cars'],
      labels= df_car_data_pie['Engine Fuel Type'].values,
      autopct= '%.2f%%',
      textprops= {'size': 'x-large',
                 'fontweight': 'bold'
                 })
py.title('Cars By Engine Type', fontsize = 15, fontweight = 'bold')
py.legend()
py.show()
```
